Para crear nuevas columnas en un dataframe de Pandas, puedes utilizar el mtodo .assign(). Este mtodo te permite crear una copia del dataframe y aadir nuevas columnas a esta copia.

Para extraer las palabras que estn a la derecha de los dos puntos (:), puedes utilizar la funcin str.split() y seleccionar el segundo elemento del resultado. Por ejemplo, para extraer la fragancia de la primera fila de tu dataframe, puedes hacer lo siguiente:

```python
df['Fragancia'] = df['columna'].str.split(':').str[1]
```

Para crear las tres nuevas columnas que has mencionado, primero puedes crear una copia del dataframe original con el mtodo .copy(), y luego aadir las nuevas columnas con el mtodo .assign(). Aqu te dejo un ejemplo de cmo podras hacerlo:

```python
df_new = df.copy()

df_new = df_new.assign(
    Fragancia = df['columna'].str.split(':').str[1],
    Aroma = df['columna'].str.split(':').str[1],
    Defecto_Atributo = df['columna'].str.split(':').str[1]
)
```
